```python
import json
import os

import pandas as pd
```

```python
os.listdir("../data")
```

```python
def load_data(file_name: str):
    with open(f"../data/{file_name}") as f:
        j = json.load(f)
    return j
    
```

```python
user_events = load_data("user_events.json")
org_events = load_data("org_events.json")
```

```python
user_df = pd.DataFrame(user_events)
org_df = pd.DataFrame(org_events)
```
